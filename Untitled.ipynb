```python
#But usually, we also want to handle the case when the condition is not true, so we use an else statement for that.
age = 15
print('Age: ' + str(age))
if age > 17:
  print('You are an adult!')
else:
  print('You are not an adult yet.')
```

```python
#We can also use else if to check more than one condition, so that we get more than two outcomes.

age = 15
print('Age: ' + str(age))

if age < 13:
  print('You are a child')
elif age < 20:
  print('You are a teenager')
else:
  print('You are an adult')
```

```python
#A nested if statement is an if statement inside another if statement.
#Nested if statements are useful in cases where you want to check a condition, only if another condition is true.

age = 19
print('Age: ' + str(age))

if age < 13:
  print('You are a child')
elif age < 20:
  print('You are a teenager')
  if age > 17:
    print('and an adult!') 
else:
  print('You are an adult')
```

```python
#Python Operators
#Operators perform operations on variables and values. Python supports several types of operators:
#1. Arithmetic Operators
#These are used for basic mathematical operations.
#Examples:
a = 10
b = 3
print(a + b)  # Addition: 13
print(a - b)  # Subtraction: 7
print(a * b)  # Multiplication: 30
print(a / b)  # Division: 3.33
print(a % b)  # Modulus (remainder): 1
print(a ** b) # Exponentiation: 1000
print(a // b) # Floor Division: 3
```

```python
#2. Assignment Operators
#Used to assign values to variables.
#Examples:
x = 5
x += 3  # x = x + 3 → 8
x *= 2  # x = x * 2 → 16
```

```python
#3. Comparison Operators
#Used to compare values and return Boolean results.
#Examples:
x = 10
y = 20
print(x == y)  # False
print(x != y)  # True
print(x < y)   # True
print(x >= y)  # False
```

```python
#4. Logical Operators
#Used to combine conditional statements.
#Examples:
x = 5
print(x > 2 and x < 10)  # True
print(x > 10 or x < 2)   # False
print(not(x > 10))       # True
```

```python
#5. Membership Operators
#Used to test if a sequence contains a value.
#Examples:
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)     # True
print("grape" not in fruits) # True

print(len("Hello"))        # Output : 5
print(abs(-10))            # Output : 10
print(sum([1, 2, 3, 4]))   # Output : 10
print(round (3.14159, 2))  # Output : 3.14
print(type(42))            # Output: <class 'int'>
```

```python
# ITERATING — Going Through Every Item
# Iterating means checking each item one by one, like checking each student's name in a list.

import numpy as np  # import numpy library

a = np.array([10, 11, 12, 13, 14])  # define a NumPy array

for i in a:  # loop through each item in the array
    print(i)
```

```python
# Slice from index 1 up to—but not including—index 5
slice1 = a[1:5]               # → [11 12 13 14]
print("a[1:5] :", slice1)

# Same slice but step by 2 (every second element)
slice2 = a[1:5:2]             # → [11 13]
print("a[1:5:2]:", slice2)
```

```python
import numpy as np            # Import NumPy

# Create an array from 10 up to (but not including) 16
a = np.arange(10, 16)         # → [10 11 12 13 14 15]
print("Full array:", a)

# Slice from index 1 up to—but not including—index 5
slice1 = a[1:5]               # → [11 12 13 14]
print("a[1:5] :", slice1)

# Same slice but step by 2 (every second element)
slice2 = a[1:5:2]             # → [11 13]
print("a[1:5:2]:", slice2)
```

```python
students = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie'],
    'Age': [20, 21, 22]
})

scores = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
})

merged = pd.merge(students, scores, on='Student')

print(merged)
```

```python
#1. Removing Duplicates
#Sometimes the same data gets entered more than once.
import pandas as pd
# Sample data with duplicates
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Email': ['alice@example.com', 'bob@example.com', 'alice@example.com', 'charlie@example.com', 'bob@example.com']
}

df = pd.DataFrame(data)

print("Before Removing Duplicates:")
print(df)
```

```python
# Remove duplicate rows
df_unique = df.drop_duplicates()

print("After Removing Duplicates:")
print(df_unique)
```

```python
#2. String Manipulation
#Let’s say you have full names in one column and want to:
#•Extract first names
#•Convert all names to title case (e.g. “john” → “John”)
data = {
    'FullName': ['john doe', 'MARY ANN', 'alice Smith']
}

df = pd.DataFrame(data)
```

```python
# Convert to title case
df['FullName'] = df['FullName'].str.title()

# Split full name into first and last
df['FirstName'] = df['FullName'].str.split().str[0]
df['LastName'] = df['FullName'].str.split().str[1]

print(df)
```
